```python
from langchain_community.llms import LlamaCpp
from langchain_core.prompts import PromptTemplate

from langchain_classic.chains import LLMChain
```

```python
llm = LlamaCpp(model_path = '/content/Phi-3-mini-4k-instruct-q4.gguf',
               max_tokens = 500,
               n_gpu_layers = -1,
               seed = 42,
               n_ctx = 2048,
               verbose = False)
```

```python
template = """<|user|>
Give the director name of the movie {movie_name}. only the name. <|end|>
<|assistant|>"""

movie_name_prompt = PromptTemplate(template = template,
                        input_variables = ['movie_name'])

director_name = LLMChain(llm = llm, prompt = movie_name_prompt, output_key = 'director_name')

director_name.invoke("interstellar")
```

```python
template = """<|user|>
give the name of the main actor casted in the movie {movie_name} directed by {director_name}. give only the name" <|end|>
<|assistant|>"""

actor_prompt = PromptTemplate(template = template,
                        input_variables = ['movie_name', 'director_name'])

actor = LLMChain(llm = llm, prompt = actor_prompt, output_key = 'actor')

llm_chain = director_name | actor

llm_chain.invoke("interstellar")
```

```python
template = """<|user|>
Describe in one line about the movie {movie_name}, which is directed by {director_name} and starred by {actor}. <|end|>
<|assistant|>"""

one_liner_prompt = PromptTemplate(template = template,
                           input_variables = ['movie_name', 'director_name', 'actor'])

one_liner = LLMChain(llm = llm, prompt = one_liner_prompt, output_key = 'one_liner')

llm_chain = director_name | actor | one_liner

llm_chain.invoke("interstellar")
```
