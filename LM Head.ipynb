```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-3-mini-4k-instruct",
                                             device_map = 'auto',
                                             torch_dtype = 'auto')
```

```python
message =[{"role": "user",
           "content" : "write a leave request mail"}]

prompt = tokenizer.apply_chat_template(message, tokenize=False, add_generation_prompt=True)

input = tokenizer(prompt, return_tensors="pt")

for k,v in input.items():
  input[k] = v.to(model.device)

output = model.generate(**input, max_new_tokens = 50, use_cache = False)

new_tokens = output[0][input['input_ids'].shape[1] : ]

response = tokenizer.decode(new_tokens, skip_special_tokens=True)
```

```python
print(f"Message : {message}\n")
print(f"Prompt : {prompt}\n")
print(f"Input : {input}\n")
print(f"Output : {output}\n")

print(f"Response : {response}\n")
```

```python
message =[{"role": "user",
           "content" : "The capital of France is"}]

prompt = tokenizer.apply_chat_template(message, tokenize=False, add_generation_prompt=True)

input = tokenizer(prompt, return_tensors="pt")

for k,v in input.items():
  input[k] = v.to(model.device)

output = model.model(**input)

lm_head_output = model.lm_head(output.last_hidden_state)


pred_ids = lm_head_output[:, -3 : ].argmax(dim=-1)
response = tokenizer.decode(pred_ids[0])
```

```python
print(f"Message : {message}\n")
print(f"Prompt : {prompt}\n")
print(f"Input : {input}\n")

for id in input['input_ids'][0]:
  print(f"{id.item()} --> {tokenizer.decode(id)}\n")

print(f"Output : {output}\n")

print(f"pred_ids : {pred_ids}\n")

print(f"response : {response}\n")


for id in pred_ids[0]:
  print(f"{id.item()} --> {tokenizer.decode(id)}\n")
```
