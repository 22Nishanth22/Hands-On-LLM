```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
```

```python
playlist_url = "https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt"
song_details_url = "https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt"
```

```python
playlist = pd.read_csv(playlist_url, sep="\t", header=None, encoding='utf-8', skiprows=2)[0]
playlist  = [p.split() for p in playlist]
song_details = pd.read_csv(song_details_url, sep="\t", header = None, names=['id', 'title','artist']).set_index('id')
```

```python
print(f"Playlist : {playlist[9]}\n")
print(f"{song_details.iloc[9]}")
```

```python
model = Word2Vec(playlist,
                 vector_size=32,
                 window = 20,
                 negative=50,
                 min_count=1,
                 workers=4)
```

```python
def recommend(id):
  similar_songs = np.array(model.wv.most_similar(positive=id, topn=5))[:,0]
  return song_details.iloc[similar_songs]
```

```python
recommend(2172)
```
